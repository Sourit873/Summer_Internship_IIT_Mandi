```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("1.0.csv")
fn1y = data['fn1y']
un1y = data['un1y']
fn2y = data['fn2y']
un2y = data['un2y']

fn1c = data['fn1c']
un1c = data['un1c']
fn2c = data['fn2c']
un2c = data['un2c']

plt.figure(figsize=(10,7))
plt.xlim(0,400)
plt.ylim(0,12)

plt.plot(un1y,fn1y,'k',label='YADE result (Glass)')
plt.plot(un1c,fn1c,'r')
plt.plot(un2y,fn2y,'b',label='YADE result (Limestone)')
plt.plot(un2c,fn2c,'r',label='Analytical solution-Hertz')

plt.legend()
plt.title("Benchmark test 1(a)")
plt.xlabel("Normal contact displacement (μm)")
plt.ylabel("Normal contact force (kN)")
#plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('4.0.csv')
theta_in = data['theta_in_y']
angVel1 = data['angVel_y1']
angVel2 = data['angVel_y2']

theta_in_c1 = data['theta_in_c1']
angVel_c1 = data['angVel_c1']
theta_in_c2 = data['theta_in_c2']
angVel_c2 = data['angVel_c2']

theta_in_a1 = data['theta_in_a1']
angVel_a = data['angVel_a']

plt.figure(figsize=(10,7))

plt.xlabel("Incident angle (degree)")

plt.xlim(0,90)
plt.ylim(-800,0)
plt.plot(theta_in_a1,angVel_a,'r',label='Analytical solution-sliding regime')
plt.plot(theta_in_c1,angVel_c1,'g',label='DEM (PFC3D) result (Al oxide)')
plt.plot(theta_in_c2,angVel_c2,'m',label='DEM (PFC3D) result (Al alloy)')

plt.plot(theta_in,angVel1,'k',label='YADE result (Al oxide)')
plt.plot(theta_in,angVel2,'b',label='YADE result (Al alloy)')

plt.legend()
plt.title("Benchmark test 4(b)")
plt.ylabel("Post-collision angular velocity (rad/s)")

#plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('4.0.csv')
theta_in = data['theta_in_y']
tan_out1 = data['tan_out1']
tan_out2 = data['tan_out2']

theta_in_1 = data['theta_in_1']
theta_out_c1 = data['theta_out_c1']
theta_in_2 = data['theta_in_2']
theta_out_c2 = data['theta_out_c2']

theta_in_a2 = data['theta_in_a2']
theta_out_a = data['theta_out_a']

theta_out1=np.degrees(np.arctan(tan_out1))
theta_out2=np.degrees(np.arctan(tan_out2))

plt.figure(figsize=(10,7))

plt.xlabel("Incident angle (degree)")

plt.xlim(0,90)
plt.ylim(-30,90)
plt.plot(theta_in_a2,theta_out_a,'r',label='Analytical solution-sliding regime')
plt.plot(theta_in_1,theta_out_c1,'g',label='DEM (PFC3D) result (Al oxide)')
plt.plot(theta_in_2,theta_out_c2,'m',label='DEM (PFC3D) result (Al alloy)')

plt.plot(theta_in,theta_out1,'k',label='YADE result (Al oxide)')
plt.plot(theta_in,theta_out2,'b',label='YADE result (Al alloy)')

plt.legend()
plt.title("Benchmark test 4(c)")
plt.ylabel("Rebound angle (degree)")

#plt.grid()
plt.show()
```
